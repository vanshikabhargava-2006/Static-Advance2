Q1. What is hypothesis testing in statistics

-> Hypothesis testing in statistics is a method used to make inferences or draw conclusions about a population
based on sample data. It involves formulating two competing hypotheses—the null hypothesis (H₀) and the
alternative hypothesis (H₁)—and using sample data to test the validity of the null hypothesis

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis

-> The null hypothesis (H₀) is a statement that there is no effect or difference in a population. It suggests that
any observed effect is due to random chance. The alternative hypothesis (H₁ or Ha) is the opposite of the null
hypothesis and posits that there is a significant effect or difference. In hypothesis testing, the goal is to
determine whether there is enough evidence to reject H₀ in favor of H₁.

Q3. What is the significance level in hypothesis testing, and why is it important

-> The significance level, often denoted as α (alpha), represents the probability of rejecting the null hypothesis
when it is actually true (Type I error). Commonly, a significance level of 0.05 is used, meaning there is a 5%
chance of incorrectly rejecting the null hypothesis. It is important because it sets the threshold for
determining statistical significance, balancing the risk of Type I and Type II errors.

Q4. What does a P-value represent in hypothesis testing

-> The P-value is the probability of observing data at least as extreme as the current sample, assuming the null
hypothesis is true. It quantifies the evidence against the null hypothesis. A smaller P-value indicates stronger
evidence to reject the null hypothesis. In most cases, if the P-value is less than the significance level (α), the
null hypothesis is rejected.




Q5. How do you interpret the P-value in hypothesis testing

->  The P-value indicates the strength of the evidence against the null hypothesis. If the P-value is small
(typically less than 0.05), it suggests that the observed data is unlikely under the null hypothesis, leading to a
rejection of H₀. If the P-value is large, it indicates that the observed data is consistent with the null hypothesis,
and H₀ is not rejected.

Q6. What are Type 1 and Type 2 errors in hypothesis testing

-> A Type 1 error occurs when the null hypothesis is incorrectly rejected (a "false positive"), meaning the test
suggests an effect when there is none. A Type 2 error occurs when the null hypothesis is not rejected when it
should be (a "false negative"), meaning the test fails to detect a real effect.

Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing

-> A one-tailed test examines whether a parameter is either greater than or less than a certain value (i.e., it
tests in one direction). A two-tailed test, on the other hand, checks for deviations in both directions, testing
whether the parameter is significantly different from a certain value in either direction.

Q8. What is the Z-test, and when is it used in hypothesis testing

-> The Z-test is a statistical test used to determine if there is a significant difference between the sample mean
and the population mean when the population variance is known. It is commonly used when the sample size
is large (n > 30) or the population variance is known

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing.
 -> The Z-score is calculated by subtracting the population mean from the sample mean and dividing by the
population standard deviation. It measures how many standard deviations the sample mean is away from
the population mean. In hypothesis testing, it helps determine how unusual or typical the sample data is
under the null hypothesis.

Q10. What is the T-distribution, and when should it be used instead of the normal distribution.

-> The T-distribution is a probability distribution that is used when the sample size is small (n < 30) and the
population variance is unknown. It is similar to the normal distribution but has heavier tails, which accounts
for the increased uncertainty with small samples.

Q11. What is the difference between a Z-test and a T-test.

-> A Z-test is used when the sample size is large and the population variance is known, while a T-test is used for
smaller sample sizes or when the population variance is unknown. The T-test uses the T-distribution, which
accounts for more variability in small samples.

Q12. What is the T-test, and how is it used in hypothesis testing.

-> The T-test is used to determine if there is a significant difference between the means of two groups or a
sample mean and a population mean. It is commonly used when the sample size is small, and the
population standard deviation is unknown. There are different types of T-tests: one-sample, independent
two-sample, and paired T-test.

Q13. What is the relationship between Z-test and T-test in hypothesis testing.

-> The Z-test and T-test both serve to compare a sample mean to a population mean, but they differ in the
circumstances under which they are used. The Z-test is preferred when the sample size is large or when the
population standard deviation is known. The T-test is used when the sample size is small or when the
population standard deviation is unknown

Q14. What is a confidence interval, and how is it used to interpret statistical results.

-> A confidence interval is a range of values that is used to estimate the true population parameter with a
certain level of confidence (e.g., 95% confidence). It provides a range of plausible values for the parameter
and indicates the uncertainty associated with the estimate

Q15. What is the margin of error, and how does it affect the confidence interval.

-> The margin of error is the amount by which the sample estimate may differ from the true population value. It
is typically calculated as a function of the standard error and the critical value for the chosen confidence
level. A larger margin of error results in a wider confidence interval, indicating greater uncertainty in the
estimate.

Q16. How is Bayes' Theorem used in statistics, and what is its significance.

-> Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It provides a
mathematical framework for revising beliefs about the probability of events in light of new data. Bayes'
Theorem is significant because it allows for continuous learning and updating of statistical models.

Q17. What is the Chi-square distribution, and when is it used

-> The Chi-square distribution is a probability distribution used in hypothesis testing, particularly in tests of
independence and goodness-of-fit. It is applied when data is categorical and helps determine if observed
frequencies differ significantly from expected frequencies.

Q18. What is the Chi-square goodness of fit test, and how is it applied

-> The Chi-square goodness of fit test is used to determine if a sample data set fits an expected probability
distribution. It compares observed frequencies with expected frequencies based on a hypothesized
distribution to assess whether the data significantly deviates from expectations.

Q19. What is the F-distribution, and when is it used in hypothesis testing

-> The F-distribution is used in hypothesis testing to compare the variances of two or more groups. It is
commonly used in analysis of variance (ANOVA) tests and regression analysis to assess if group variances
are equal.

Q20. What is an ANOVA test, and what are its assumptions

-> ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups. Its
assumptions include normality (the data is approximately normally distributed), independence of
observations, and homogeneity of variances (the variances in the groups are equal).

Q21. What are the different types of ANOVA tests

-> The main types of ANOVA tests are one-way ANOVA (comparing means across one factor with more than
two groups), two-way ANOVA (comparing means across two factors), and repeated measures ANOVA
(comparing means when the same subjects are used for each condition).

Q22. What is the F-test, and how does it relate to hypothesis testing

->  The F-test is used to compare the variances of two or more groups to assess whether they come from
populations with the same variance. It is commonly used in ANOVA and regression analysis to determine if
the model explains a significant amount of variation in the data

```python
#Write a Python program to perform a Z-test for comparing a sample mean to aknown population mean and interpret the results

import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, sample_size):
    # Calculate sample mean
    sample_mean = np.mean(sample_data)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    return z_statistic, p_value

# Example usage
# Sample data
sample_data = [15, 16, 14, 15, 16, 17, 16, 14, 15, 16]
population_mean = 15  # Known population mean
population_std = 1.5  # Known population standard deviation
sample_size = len(sample_data)

# Perform Z-test
z_stat, p_value = z_test(sample_data, population_mean, population_std, sample_size)

# Print results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret results
alpha = 0.05  # Significance level (5%)

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

```python
# Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample_data, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate sample mean
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Critical Z value for two-tailed test at significance level alpha
    critical_value = stats.norm.ppf(1 - alpha/2)

    # Decision region: Accept H0 if z_statistic lies within the critical region
    reject_null = abs(z_statistic) > critical_value

    return z_statistic, p_value, reject_null, critical_value

# Example usage
# Sample data
sample_data = [22, 25, 28, 24, 26, 23, 29, 24, 25, 27]
population_mean = 25  # Known population mean
population_std = 2  # Known population standard deviation
sample_size = len(sample_data)

# Perform Two-Tailed Z-test
z_stat, p_value, reject_null, critical_value = two_tailed_z_test(sample_data, population_mean, population_std, sample_size)

# Print results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-value: ±{critical_value:.4f}")
if reject_null:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")

# Plotting the normal distribution and decision region
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label='Rejection Region')
plt.axvline(x=z_stat, color='green', linestyle='--', label=f'Z-statistic = {z_stat:.4f}')

# Adding labels and title
plt.title('Two-Tailed Z-Test: Decision Region')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
#Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_type_1_and_type_2_errors(population_mean, population_std, sample_size, alpha=0.05, alternative_mean=None):
    """
    This function calculates Type 1 and Type 2 errors for a hypothesis test, and visualizes them.

    :param population_mean: Known population mean (H0)
    :param population_std: Known population standard deviation
    :param sample_size: Size of the sample
    :param alpha: Significance level (default 0.05)
    :param alternative_mean: Mean of the alternative hypothesis (H1), None for Type 1 error
    :return: Type 1 and Type 2 error rates, and visualization
    """

    # Critical Z-value for two-tailed test at significance level alpha
    critical_value = stats.norm.ppf(1 - alpha / 2)

    # Simulate the scenario for Type 1 Error (H0 is true, we reject it)
    # Type 1 Error: H0 is true, but we reject it.
    z_critical_low = -critical_value
    z_critical_high = critical_value

    # Simulate the scenario for Type 2 Error (H1 is true, we fail to reject H0)
    # Type 2 Error: H1 is true, but we fail to reject H0.
    # Assuming the true population mean is alternative_mean
    if alternative_mean is None:
        alternative_mean = population_mean + 2  # Example: shift mean by 2 for Type 2 error

    # Calculate the Type 2 error (area where we fail to reject H0 when H1 is true)
    beta = stats.norm.cdf(z_critical_low, loc=alternative_mean, scale=population_std / np.sqrt(sample_size)) + \
           (1 - stats.norm.cdf(z_critical_high, loc=alternative_mean, scale=population_std / np.sqrt(sample_size)))

    # Calculate the Type 1 error (alpha)
    type_1_error = alpha

    # Visualize the regions
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="Standard Normal Distribution", color="blue")

    # Type 1 error (rejection region when H0 is true)
    plt.fill_between(x, y, where=(x <= z_critical_low) | (x >= z_critical_high), color="red", alpha=0.5, label="Type 1 Error (α)")

    # Type 2 error (failure to reject H0 when H1 is true)
    x_alt = np.linspace(-4, 4, 1000)
    y_alt = stats.norm.pdf(x_alt, loc=alternative_mean, scale=population_std / np.sqrt(sample_size))
    plt.fill_between(x_alt, y_alt, where=(x_alt >= z_critical_low) & (x_alt <= z_critical_high), color="yellow", alpha=0.5, label="Type 2 Error (β)")

    # Mark critical regions and acceptance regions
    plt.axvline(x=z_critical_low, color="green", linestyle="--", label=f"Critical Z = ±{critical_value:.2f}")
    plt.axvline(x=z_critical_high, color="green", linestyle="--")

    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)

    # Show plot
    plt.show()

    return type_1_error, beta

# Example usage
population_mean = 50
population_std = 10
sample_size = 30
alpha = 0.05
alternative_mean = 52  # Alternative hypothesis mean (H1) for Type 2 error simulation

# Calculate Type 1 and Type 2 errors and visualize the regions
type_1_error, type_2_error = calculate_type_1_and_type_2_errors(population_mean, population_std, sample_size, alpha, alternative_mean)

print(f"Type 1 Error (α): {type_1_error:.4f}")
print(f"Type 2 Error (β): {type_2_error:.4f}")
```

```python
#Write a Python program to perform an independent T-test and interpret the results
import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    # Calculate the means of the samples
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)

    # Calculate the standard deviations of the samples
    std1 = np.std(sample1, ddof=1)  # Sample standard deviation with Bessel's correction
    std2 = np.std(sample2, ddof=1)

    # Calculate the sample sizes
    n1 = len(sample1)
    n2 = len(sample2)

    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))

    # Calculate the t-statistic
    t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(1/n1 + 1/n2))

    # Calculate degrees of freedom
    degrees_of_freedom = n1 + n2 - 2

    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_of_freedom))

    return t_statistic, p_value, degrees_of_freedom

# Example usage
# Sample data
sample1 = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
sample2 = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65]

# Perform independent T-test
t_stat, p_value, df = independent_t_test(sample1, sample2)

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {df}")

# Interpret results
alpha = 0.05  # Significance level (5%)

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two sample means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two sample means.")
```

```python
#Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    sample_data (list or array): The sample data.
    confidence_level (float): The confidence level (default is 95%).

    Returns:
    tuple: Lower and upper bounds of the confidence interval.
    """
    # Sample size
    n = len(sample_data)

    # Sample mean
    sample_mean = np.mean(sample_data)

    # Sample standard deviation (Bessel's correction applied)
    sample_std = np.std(sample_data, ddof=1)

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(n)

    # Confidence level z-score (for large samples, we use z-distribution; for smaller samples, t-distribution)
    alpha = 1 - confidence_level
    critical_value = stats.t.ppf(1 - alpha/2, df=n-1)  # For T-distribution (use Z-distribution for large samples)

    # Margin of error
    margin_of_error = critical_value * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
confidence_level = 0.95  # 95% confidence level

# Calculate the confidence interval
lower, upper = confidence_interval(sample_data, confidence_level)

print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")

```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayesian_inference(prior, likelihood, false_positive_rate, test_accuracy):
    """
    Perform Bayesian inference using Bayes' Theorem.

    Parameters:
    prior (float): Prior probability of the hypothesis (probability of having the disease).
    likelihood (float): Likelihood of the evidence given the hypothesis (probability of a positive test given the disease).
    false_positive_rate (float): Likelihood of the evidence given the negation of the hypothesis (false positive rate).
    test_accuracy (float): The probability of a positive test regardless of disease status (marginal likelihood).

    Returns:
    float: Posterior probability of the hypothesis given the evidence (probability of having the disease given a positive test).
    """

    # The marginal likelihood P(E) is computed as follows:
    marginal_likelihood = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Apply Bayes' Theorem
    posterior = (likelihood * prior) / marginal_likelihood
    return posterior

# Example usage
prior = 0.01  # 1% prior probability of having the disease
likelihood = 0.95  # 95% chance of a positive test if you have the disease
false_positive_rate = 0.05  # 5% false positive rate
test_accuracy = 0.05  # This is used for calculating P(E), but not directly needed here.

# Calculate the posterior probability
posterior = bayesian_inference(prior, likelihood, false_positive_rate, test_accuracy)

print(f"The probability of having the disease given a positive test result is: {posterior:.4f}")
```

```python
#Write a Python program to calculate the expected frequencies for a Chi-square test based on observed dataD
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    observed_data (2D array): The contingency table of observed frequencies.

    Returns:
    2D array: The contingency table of expected frequencies.
    """
    # Step 1: Calculate the row totals and column totals
    row_totals = np.sum(observed_data, axis=1, keepdims=True)
    column_totals = np.sum(observed_data, axis=0, keepdims=True)

    # Step 2: Calculate the total number of observations
    total = np.sum(observed_data)

    # Step 3: Calculate the expected frequencies using the formula
    expected_data = (row_totals @ column_totals) / total

    return expected_data

# Example observed data (2x2 contingency table)
observed_data = np.array([[30, 10],  # Male: 30 Yes, 10 No
                          [20, 20]]) # Female: 20 Yes, 20 No

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Data:\n", observed_data)
print("\nExpected Frequencies:\n", expected_frequencies)
```

```python
#Implement an F-test using Python to compare the variances of two random samplesD
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    sample1 (array): First random sample.
    sample2 (array): Second random sample.
    alpha (float): Significance level, default is 0.05.

    Returns:
    None
    """
    # Calculate the variances of the two samples
    var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
    var2 = np.var(sample2, ddof=1)  # Sample variance for sample2

    # Calculate the F-statistic (variance ratio)
    if var1 > var2:
        F_statistic = var1 / var2  # Larger variance in the numerator
    else:
        F_statistic = var2 / var1  # Larger variance in the numerator

    # Degrees of freedom for both samples
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Perform the F-test using the cumulative distribution function (CDF)
    p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

    # Display the results
    print(f"F-statistic: {F_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation of the result
    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Example usage: Perform F-test on two random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(0, 2, 100)  # Sample 1 from normal distribution with variance 2
sample2 = np.random.normal(0, 3, 100)  # Sample 2 from normal distribution with variance 3

# Perform the F-test
f_test(sample1, sample2, alpha=0.05)
```

```python
#Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    """
    Check the assumptions for ANOVA: normality, independence, and equal variances.

    Parameters:
    groups (list of arrays): A list containing the different groups of data to be compared.

    Returns:
    None
    """
    # Assumption 1: Normality check (Shapiro-Wilk test for normality)
    print("\nChecking normality assumption for each group:")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1} - Shapiro-Wilk test: Stat={stat:.4f}, P-value={p_value:.4f}")
        if p_value < 0.05:
            print(f"  Group {i+1} does not follow normal distribution (Reject H0).")
        else:
            print(f"  Group {i+1} follows normal distribution (Fail to reject H0).")

    # Assumption 2: Independence
    print("\nIndependence assumption: This assumption is assumed to be satisfied based on the study design.")

    # Assumption 3: Equal variance (Homogeneity of variances) using Levene's test
    print("\nChecking homogeneity of variances using Levene's test:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test: Stat={stat:.4f}, P-value={p_value:.4f}")
    if p_value < 0.05:
        print("  Variances are significantly different (Reject H0). Homogeneity of variances assumption is violated.")
    else:
        print("  Variances are not significantly different (Fail to reject H0). Homogeneity of variances assumption is satisfied.")

    # Visualization: Boxplot to visually check the equal variance assumption
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=groups)
    plt.title("Boxplot to check equal variances assumption")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.show()

# Example usage: Simulating multiple groups of data for assumption checking
np.random.seed(42)  # For reproducibility

# Simulating 3 groups of data (Group1, Group2, Group3)
group1 = np.random.normal(50, 10, 30)  # Group 1: mean = 50, std = 10, n = 30
group2 = np.random.normal(55, 10, 30)  # Group 2: mean = 55, std = 10, n = 30
group3 = np.random.normal(60, 10, 30)  # Group 3: mean = 60, std = 10, n = 30

# Check the assumptions for ANOVA
check_anova_assumptions([group1, group2, group3])
```

```python
# Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

def perform_two_way_anova_and_plot():
    """
    Perform a two-way ANOVA test to study the interaction between two factors
    and visualize the results.
    """
    # Simulating data for the two-way ANOVA
    np.random.seed(42)  # For reproducibility

    # Factor A: Treatment (3 levels: A1, A2, A3)
    treatment = np.tile(["A1", "A2", "A3"], 30)

    # Factor B: Time (2 levels: T1, T2)
    time = np.repeat(["T1", "T2"], 45)

    # Dependent variable: Outcome variable (normally distributed)
    outcome = np.random.normal(50, 10, 90)

    # Adding some variation to the dependent variable based on factors
    outcome[:30] += 5  # A1 group has a slightly higher outcome
    outcome[30:60] -= 3  # A2 group has a lower outcome
    outcome[60:] += 10  # A3 group has a higher outcome

    # Create a DataFrame for the data
    data = pd.DataFrame({
        'Treatment': treatment,
        'Time': time,
        'Outcome': outcome
    })

    # Fit the two-way ANOVA model
    model = smf.ols('Outcome ~ C(Treatment) + C(Time) + C(Treatment):C(Time)', data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)

    # Display the ANOVA table
    print(anova_table)

    # Interpretation of the results
    if anova_table['PR(>F)'][0] < 0.05:
        print("Treatment factor has a significant effect on the outcome.")
    if anova_table['PR(>F)'][1] < 0.05:
        print("Time factor has a significant effect on the outcome.")
    if anova_table['PR(>F)'][2] < 0.05:
        print("There is a significant interaction effect between Treatment and Time.")

    # Visualize the results
    plt.figure(figsize=(10, 6))
    sns.pointplot(x="Treatment", y="Outcome", hue="Time", data=data,
                  dodge=True, markers=["o", "s"], linestyles=["-", "--"])
    plt.title('Two-Way ANOVA: Interaction between Treatment and Time', fontsize=14)
    plt.xlabel('Treatment', fontsize=12)
    plt.ylabel('Outcome', fontsize=12)
    plt.legend(title='Time')
    plt.show()

# Run the function
perform_two_way_anova_and_plot()

```

```python
#D Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_f_distribution(df1, df2, alpha=0.05):
    """
    Visualizes the F-distribution for given degrees of freedom and marks the critical value.

    Parameters:
    df1 (int): Degrees of freedom for the numerator (between-group variance).
    df2 (int): Degrees of freedom for the denominator (within-group variance).
    alpha (float): Significance level for the hypothesis test (default is 0.05).
    """
    # Generate F-distribution values for plotting
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, df1, df2)

    # Plot the F-distribution curve
    plt.plot(x, y, label=f'F-distribution df1={df1}, df2={df2}', color='blue', lw=2)

    # Find the critical F value for the given significance level (alpha)
    critical_value = stats.f.ppf(1 - alpha, df1, df2)

    # Highlight the rejection region
    plt.fill_between(x, 0, y, where=(x >= critical_value), color='red', alpha=0.3, label=f'Rejection Region (F > {critical_value:.2f})')

    # Annotate the critical value on the plot
    plt.axvline(x=critical_value, color='red', linestyle='--', label=f'Critical Value (F = {critical_value:.2f})')

    # Add labels and title
    plt.title('F-distribution with Degrees of Freedom', fontsize=14)
    plt.xlabel('F-statistic', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.legend(loc='upper right')
    plt.grid(True)

    # Show the plot
    plt.show()

# Example usage: Visualizing the F-distribution with df1=5 and df2=10
visualize_f_distribution(df1=5, df2=10)
```

```python
#Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def perform_one_way_anova_and_visualize():
    """
    Perform a One-Way ANOVA test to compare means between multiple groups
    and visualize the results with boxplots.
    """
    # Simulating data for three groups
    np.random.seed(42)

    # Group 1: Treatment A
    group1 = np.random.normal(50, 10, 30)  # mean=50, std=10, sample size=30

    # Group 2: Treatment B
    group2 = np.random.normal(55, 10, 30)  # mean=55, std=10, sample size=30

    # Group 3: Treatment C
    group3 = np.random.normal(60, 10, 30)  # mean=60, std=10, sample size=30

    # Perform the one-way ANOVA
    f_statistic, p_value = stats.f_oneway(group1, group2, group3)

    # Print ANOVA results
    print(f"ANOVA test results:")
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the p-value
    if p_value < 0.05:
        print("There is a significant difference between the group means.")
    else:
        print("There is no significant difference between the group means.")

    # Prepare data for visualization
    data = [group1, group2, group3]
    labels = ['Treatment A', 'Treatment B', 'Treatment C']

    # Create a boxplot to visualize group comparisons
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=data, notch=True, vert=True, patch_artist=True)
    plt.xticks(ticks=np.arange(3), labels=labels)
    plt.title('One-Way ANOVA: Comparison of Treatment Groups', fontsize=14)
    plt.xlabel('Treatment Group', fontsize=12)
    plt.ylabel('Outcome', fontsize=12)
    plt.grid(True)
    plt.show()

# Run the function
perform_one_way_anova_and_visualize()
```

```python
# Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def simulate_and_test_means():
    """
    Simulate random data from a normal distribution, perform hypothesis testing
    (two-sample t-test) to evaluate the means, and interpret the results.
    """
    # Set seed for reproducibility
    np.random.seed(42)

    # Simulate two random samples from normal distributions
    # Sample 1: Mean = 50, Standard Deviation = 10, Sample Size = 30
    sample1 = np.random.normal(50, 10, 30)

    # Sample 2: Mean = 55, Standard Deviation = 10, Sample Size = 30
    sample2 = np.random.normal(55, 10, 30)

    # Perform a two-sample t-test (assuming unequal variances)
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    # Print t-test results
    print(f"Two-Sample t-test results:")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the results
    if p_value < 0.05:
        print("We reject the null hypothesis: The means of the two samples are significantly different.")
    else:
        print("We fail to reject the null hypothesis: There is no significant difference between the means of the two samples.")

    # Visualize the data (boxplot and histogram)
    plt.figure(figsize=(12, 6))

    # Boxplot
    plt.subplot(1, 2, 1)
    plt.boxplot([sample1, sample2], labels=["Sample 1", "Sample 2"], patch_artist=True)
    plt.title('Boxplot of Sample Data')

    # Histogram
    plt.subplot(1, 2, 2)
    plt.hist(sample1, bins=10, alpha=0.7, label='Sample 1', color='blue', edgecolor='black')
    plt.hist(sample2, bins=10, alpha=0.7, label='Sample 2', color='green', edgecolor='black')
    plt.legend()
    plt.title('Histogram of Sample Data')

    plt.tight_layout()
    plt.show()

# Run the function to simulate data and perform hypothesis testing
simulate_and_test_means()
```

```python
#Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

def hypothesis_test_variance(population_variance, sample_data):
    """
    Perform a hypothesis test for population variance using the Chi-square distribution.

    Parameters:
    - population_variance (float): The hypothesized population variance (σ₀²).
    - sample_data (array): The sample data to test.

    Returns:
    - None: Prints the results of the hypothesis test.
    """
    # Sample size (n) and sample variance (s²)
    n = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for sample variance)

    # Chi-square test statistic
    chi_square_statistic = (n - 1) * sample_variance / population_variance

    # Degrees of freedom (df)
    df = n - 1

    # Compute p-value for the two-tailed test
    p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df), 1 - stats.chi2.cdf(chi_square_statistic, df))

    # Print the results
    print(f"Chi-square test statistic: {chi_square_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation based on a 0.05 significance level
    if p_value < 0.05:
        print("We reject the null hypothesis: The sample variance is significantly different from the population variance.")
    else:
        print("We fail to reject the null hypothesis: The sample variance is not significantly different from the population variance.")

# Simulate sample data
np.random.seed(42)
sample_data = np.random.normal(100, 15, 30)  # Mean=100, SD=15, n=30

# Population variance under null hypothesis (σ₀²)
population_variance = 225  # Assume the population variance is 15^2 = 225

# Perform hypothesis test for variance
hypothesis_test_variance(population_variance, sample_data)
```

```python
#Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

def z_test_for_proportions(successes1, size1, successes2, size2):
    """
    Perform a Z-test for comparing proportions between two datasets or groups.

    Parameters:
    - successes1: Number of successes in the first sample/group.
    - size1: Sample size of the first sample/group.
    - successes2: Number of successes in the second sample/group.
    - size2: Sample size of the second sample/group.

    Returns:
    - None: Prints the results of the Z-test.
    """
    # Calculate sample proportions
    p1 = successes1 / size1
    p2 = successes2 / size2

    # Pooled proportion
    P = (successes1 + successes2) / (size1 + size2)

    # Standard error of the difference in proportions
    SE = np.sqrt(P * (1 - P) * (1 / size1 + 1 / size2))

    # Z-statistic
    Z = (p1 - p2) / SE

    # Two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

    # Print the results
    print(f"Z-statistic: {Z:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation based on a 0.05 significance level
    if p_value < 0.05:
        print("We reject the null hypothesis: The proportions of the two groups are significantly different.")
    else:
        print("We fail to reject the null hypothesis: There is no significant difference between the proportions.")

# Example data
successes1 = 120  # Number of successes in group 1
size1 = 200       # Sample size of group 1
successes2 = 150  # Number of successes in group 2
size2 = 250       # Sample size of group 2

# Perform the Z-test for proportions
z_test_for_proportions(successes1, size1, successes2, size2)
```

```python
# Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_for_variance(data1, data2):
    """
    Perform an F-test to compare the variances of two datasets.

    Parameters:
    - data1: The first dataset.
    - data2: The second dataset.

    Returns:
    - None: Prints the results of the F-test.
    """
    # Sample sizes
    n1 = len(data1)
    n2 = len(data2)

    # Sample variances
    s1_squared = np.var(data1, ddof=1)  # Sample variance for data1
    s2_squared = np.var(data2, ddof=1)  # Sample variance for data2

    # F-statistic (ratio of variances)
    f_statistic = s1_squared / s2_squared if s1_squared > s2_squared else s2_squared / s1_squared

    # Degrees of freedom
    df1 = n1 - 1
    df2 = n2 - 1

    # P-value from the F-distribution (two-tailed test)
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    # Print the results
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation based on a 0.05 significance level
    if p_value < 0.05:
        print("We reject the null hypothesis: The variances of the two datasets are significantly different.")
    else:
        print("We fail to reject the null hypothesis: The variances of the two datasets are not significantly different.")

    # Visualization: F-distribution and observed F-statistic
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, df1, df2)

    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})')
    plt.axvline(f_statistic, color='r', linestyle='--', label=f'Observed F-statistic: {f_statistic:.4f}')
    plt.title('F-Test for Comparing Variances')
    plt.xlabel('F-statistic')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

# Example data: Two samples with different variances
np.random.seed(42)
data1 = np.random.normal(loc=0, scale=5, size=50)  # Sample 1 with variance 25
data2 = np.random.normal(loc=0, scale=10, size=50)  # Sample 2 with variance 100

# Perform the F-test for variances
f_test_for_variance(data1, data2)
```

```python
#Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_goodness_of_fit(observed, expected):
    """
    Perform a Chi-square goodness of fit test.

    Parameters:
    - observed: The observed frequencies (data).
    - expected: The expected frequencies based on the null hypothesis.

    Returns:
    - None: Prints the results of the Chi-square test.
    """
    # Calculate Chi-square statistic and p-value
    chi_square_statistic, p_value = stats.chisquare(observed, expected)

    # Print the results
    print(f"Chi-square statistic: {chi_square_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation based on a 0.05 significance level
    if p_value < 0.05:
        print("We reject the null hypothesis: The observed data does not fit the expected distribution.")
    else:
        print("We fail to reject the null hypothesis: The observed data fits the expected distribution.")

    # Visualization: Bar plot of observed vs expected frequencies
    categories = np.arange(1, len(observed) + 1)

    plt.bar(categories - 0.2, observed, width=0.4, label='Observed', color='blue', align='center')
    plt.bar(categories + 0.2, expected, width=0.4, label='Expected', color='red', align='center')

    plt.xlabel('Categories')
    plt.ylabel('Frequency')
    plt.title('Chi-square Goodness of Fit Test')
    plt.legend()
    plt.show()

# Simulate data
np.random.seed(42)
# Suppose we are testing if a die is fair: Expected frequencies for each face of a die (1-6)
expected_frequencies = np.array([1/6] * 6) * 100  # Expected 100 rolls of a fair die

# Simulate 100 rolls of a fair die (with some random variance)
observed_frequencies = np.random.multinomial(100, [1/6] * 6)

# Perform Chi-square goodness of fit test
chi_square_goodness_of_fit(observed_frequencies, expected_frequencies)
```
